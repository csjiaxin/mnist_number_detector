```python
# single linear, without hidden layers
# Cell 1: Imports and setup
from common_imports import *
from common_utils import evaluate, train_model, setup_data

class SimpleMLP(nn.Module):
    def __init__(self, in_features=28*28, hidden_sizes=(256, 128), num_classes=10, dropout=0.2):
        super().__init__()
      
       
        self.net = nn.Sequential(
           nn.Linear(in_features, num_classes)
        )

    def forward(self, x):
        x = x.view(x.size(0), -1)
        return self.net(x)
    
# 4) Data setup (Train/Validation/Test)
BATCH_SIZE = 32
NUM_EPOCHS = 1
LEARNING_RATE = 1e-3
VAL_RATIO = 0.1

train_loader, val_loader, test_loader, device = setup_data(batch_size=BATCH_SIZE, val_ratio=VAL_RATIO)

# 5) Train and evaluate
model = SimpleMLP().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)

history = train_model(
    model,
    train_loader,
    val_loader,
    criterion,
    optimizer,
    device,
    num_epochs=NUM_EPOCHS,
    patience=2,
)

test_loss, test_acc = evaluate(model, test_loader, criterion, device)
print(f"\nFinal Test Loss: {test_loss:.4f} | Final Test Accuracy: {test_acc:.4f}")
```
